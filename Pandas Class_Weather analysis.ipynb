### Python loc() function – Extract values from a dataset

####  Working of Python loc() function
#### Python comprises various modules that have in-built functions to deal with and manipulate the data values.

####  The . loc property of the DataFrame object allows the return of specified rows and/or columns from that DataFrame . It can do so using a label or label(s), or a boolean array of the same size as the axis being filtered.

#### What is iloc() Function in Python? iloc stands for “integer location” and is a function in the Pandas library. It is used to select rows and columns from a Pandas DataFrame or a Series using integer-based indexing.

```python
weather_df.loc[0:5, ['Visibility (km)','Rel Hum (%)']]
```

```python
weather_df[['Visibility (km)','Rel Hum (%)']].loc[0:5]
```

```python
weather_df[['Rel Hum (%)','Visibility (km)']].iloc[:5:2]
```
